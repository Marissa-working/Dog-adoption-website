# 1. yelp dataset -> get the pet store/pet services/veterinarians

```python
import pandas as pd

df = pd.read_json('yelp_academic_dataset_business.json', lines=True)
```

```python
df.head()
```

```python
# There is only 27 states in the dataset, and AB is in the Canada, XMS, VI is in the United Kingdom
# So 24 states data can be used
states = df["state"].unique()
states
```

```python
# drop the places outside the United States
df = df[df["state"] != 'AB'] # AB -> Canada: AB (Alberta)
df = df[df["state"] != 'XMS'] # XMS -> United Kingdom
df = df[df["state"] != 'VI'] # VI ->  United Kingdom
# drop the NA in categories column
df = df.dropna(subset=['categories'])
# filter the pet stores, dog parks from the categories
data = df[df["categories"].str.contains("Pet")]
# get the selected column in the dataset
data = data[["business_id", "name","categories","city","state","postal_code","stars"]]
clinics = data[data["categories"].str.contains("Veterinarians")]
pet_care = data[data["categories"].str.contains("Pet Services")]
pet_store = data[data["categories"].str.contains("Pet Stores")]

# export the Veterinarians dataset
clinics.to_csv('clinics.csv', index=False)
# export the pet_care dataset
pet_care.to_csv('pet_care.csv', index=False)
# export the pet_store dataset
pet_store.to_csv('pet_store.csv', index=False)
```
